```python
import os

import pandas as pd
from pandas import DataFrame

path: str = (r'C:\nihal\PROJECT\rough\playpen-paper-2025\SFT\Data\final_data\all_games_merged_old_processed.jsonl')
data: DataFrame = pd.read_json(path, lines=True)

```

```python
games = [
    'wordle',
    'wordle_withclue',
    'wordle_withcritic',
    'taboo',
    'imagegame',
    'privateshared',
    'referencegame'
]

data_successful = data[data.Success == 1]
data_successful.chat = data_successful.chat.apply(ensure_alternating_roles)
```

```python
def prepare_qa_data(data: DataFrame) -> DataFrame:
    result_data: pd.DataFrame = {key: [] for key in data.columns}

    for i, row in data.iterrows():
        chat = row['chat']
        game = row['game']

        chat_last_index = len(chat)

        for i, item in enumerate(chat):
            if item['role'] == 'assistant':
                for key in data.columns:
                    if key != 'chat':
                        result_data[key].append(row[key])
                    else:
                        if i + 1 < chat_last_index:
                            result_data[key].append(chat[:i + 1])
                        else:
                            result_data[key].append(row[key])
    return result_data
```

```python
data_successful  = data_successful.sample(frac = 1)

qa_data = prepare_qa_data(data_successful)
data_collapsed = pd.DataFrame(qa_data)

len(data_successful), len(data_collapsed)
```

```python


for idx, game in enumerate(games):
    leave_one_out = data_collapsed[data_collapsed['game'] != game]
    print(len(leave_one_out), game)
    save_path: str = fr'C:\nihal\PROJECT\rough\playpen-paper-2025\SFT\Data\final_data\D6000{idx+1}.csv'
    leave_one_out.to_csv(save_path)
```
